## **Your Task**

**What is the number of runic symbols engraved on the shield that form runic words?**

````python
import re

from more_itertools import flatten

from test_utilities import test

tests = [
    {
        "name": "Example",
        "s": """
            WORDS:THE,OWE,MES,ROD,HER,QAQ

            AWAKEN THE POWE ADORNED WITH THE FLAMES BRIGHT IRE
            THE FLAME SHIELDED THE HEART OF THE KINGS
            POWE PO WER P OWE R
            THERE IS THE END
            QAQAQ
        """,
        "expected": 42,
    },
]


@test(tests=tests)
def part_II(s: str) -> int:
    words, _, *sentences = s.strip().splitlines()
    words = set(flatten((w, w[::-1]) for w in words.removeprefix("WORDS:").split(",")))
    runes = [re.compile(rf"(?=({w}))") for w in words]
    sentences = " ".join(se.strip() for se in sentences)

    indici = set()
    for rune in runes:
        indici |= set(
            flatten(
                range(m.start(), m.start() + len(m.group(1)))
                for m in rune.finditer(sentences)
            )
        )
    return len(indici)
````

# **Quest 2: The Runes of Power**

## **Part III**

You advance carefully to the final examination chamber, where a set of scale armour awaits the candidates. This armour, unlike anything you've encountered before, is covered in scales, each etched with a runic symbol. The scales are meticulously arranged in a **rectangular grid**, with the **right edge seamlessly connected to the left**, forming an **unbroken loop**.

In this challenge, neighbouring scales can form runic words **both horizontally and vertically**, in all four directions:

- left → right
- right → left
- top → bottom
- bottom → top

Your task remains the same as before: **find all the runic words hidden within the armour and count the number of scales that compose them.**

---

## **Example**

### **WORDS:**


### **Final Count:**

Counting all highlighted scales gives:

## **Your Task**

**What is the number of scales on the armour that form runic words?**

---


````python
from functools import reduce
from itertools import takewhile
import re


from util import Str
from test_utilities import test

tests = [
    {
        "name": "Example 1",
        "s": """
            WORDS:THE,OWE,MES,ROD,RODEO

            HELWORLT
            ENIGWDXL
            TRODEOAL
        """,
        "expected": 10,
    },
]


class ScaleArmour(Str):
    def __init__(self, s: str) -> None:
        words, grid = re.split(r"(?:\r?\n\s*\r?\n)+", s.strip())

        self.runes = [
            (len(w), re.compile(rf"(?=({w[::i]}))"))
            for w in words.removeprefix("WORDS:").split(",")
            for i in (1, -1)
        ]
        self.grid = [l.strip() for l in grid.splitlines()]
        self.grid_T = ["".join(c) for c in zip(*self.grid)]

    def runes_in_row(
        self, r: int, grid: list[str], transposed: bool = False
    ) -> set[tuple[int, int]]:

        row = grid[r] if transposed else 2 * grid[r]
        n = len(grid[r])

        return {
            (c % n, r) if transposed else (r, c % n)
            for length, rune in self.runes
            for m in takewhile(lambda m: m.start() < n, rune.finditer(row))
            for c in range(m.start(), m.start() + length)
        }

    def rues_in_grid(
        self, grid: list[str], transposed: bool = False
    ) -> set[tuple[int, int]]:
        return reduce(
            set.union,
            (self.runes_in_row(r, grid, transposed) for r in range(len(grid))),
            set(),
        )

    def scales_forming_runes(self) -> int:
        return len(self.rues_in_grid(self.grid) | self.rues_in_grid(self.grid_T, True))


@test(tests=tests)
def part_III(s: str) -> int:
    sa = ScaleArmour(s)
    return sa.scales_forming_runes()
````

````python
with open("../inputs/everybody_codes_e2024_q02_p3.txt") as f:
    notes3 = f.read()

print(
    f"The exact number of potions you need to order for the final round: {part_III(notes3)}"
)
````
