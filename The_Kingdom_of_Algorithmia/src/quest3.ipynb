## **Part I**

The top knights of the Golden Duck Order are renowned not only for their skill in battle, but also for their mastery of construction. They lead all royal building projects, ensuring their designs are both impressive and environmentally friendly. Their structures blend harmoniously with nature, and their methods are praised for protecting the land throughout the building process.

In the Kingdom of Algorithmia, mining for gold, silver, and other resources follows a unique approach. Alchemists inspect the land and mark promising areas on a grid map (_your notes_) using the **`#`** symbol. Once marked, mining may begin — but it must be done with great care. The **height difference between any two adjacent areas** (side‑to‑side or top‑to‑bottom) **must never exceed one unit**. And of course, digging is only allowed in the squares marked with **`#`**.

After the digging is complete, water mages fill the excavated spaces with water, creating peaceful ponds that soon become thriving habitats. The survival of these ponds depends entirely on maintaining the correct slope during excavation. A single mistake — a slope steeper than allowed — could destabilize the terrain, disrupt mining, or create dangerous drops in the pond.

For this task, imagine the land as a flat grid, where each square represents a **1×1 unit** of earth. The digging machine removes earth **one cubic unit at a time**.  
Your goal is to determine **the maximum number of earth blocks that can be safely removed** from the marked area while keeping the slope within the allowed limits.

## **Your Task**

**How many earth blocks can be safely removed at most from your map while maintaining the proper slope?**


````python
from collections.abc import Iterator
from itertools import product

from test_utilities import test

tests = [
    {
        "name": "Example 1",
        "s": """
            ..........
            ..###.##..
            ...####...
            ..######..
            ..######..
            ...####...
            ..........
        """,
        "expected": 35,
    },
]


class Grid:
    DRDC = (-1, 0), (0, 1), (1, 0), (0, -1)
    MAX_NEIGBORS = 4

    def __init__(self, s: str) -> None:
        self.grid = [r.strip() for r in s.strip().splitlines()]

    def dig(self, debug=False) -> int:
        rows, cols = len(self.grid), len(self.grid[0])

        cubes = {
            (r, c): 1
            for r, c in product(range(rows), range(cols))
            if self.grid[r][c] == "#"
        }

        internals = {
            (r, c)
            for r, c in cubes.keys()
            if self.neigbor_count(r, c, set(cubes)) == self.MAX_NEIGBORS
        }

        while internals:
            for r, c in internals:
                cubes[(r, c)] += 1

            internals = {
                (r, c)
                for r, c in cubes.keys()
                if self.neigbor_count(r, c, internals) == self.MAX_NEIGBORS
            }

        if debug:
            print(self.debug(cubes))

        return sum(cubes.values())

    def neigbor_count(self, row: int, col: int, cubes: set[tuple[int, int]]) -> int:
        return sum(1 for dr, dc in self.DRDC if (row + dr, col + dc) in cubes)

    def debug(self, cubes: dict[tuple[int, int], int]) -> str:
        rows, cols = len(self.grid), len(self.grid[0])

        return "\n".join(
            "".join(f"{cubes.get((r,c), '.')}" for c in range(cols))
            for r in range(rows)
        )

    def __str__(self) -> str:
        return "\n".join(r for r in self.grid)


@test(tests=tests)
def part_I(s: str) -> int:
    g = Grid(s)
    return g.dig()
````

````python
with open("../inputs/everybody_codes_e2024_q03_p1.txt") as f:
    notes1 = f.read()


print(
    f"The is the exact number of potions that need to be prepared for my battle: {part_I(notes1)}"
)
````
