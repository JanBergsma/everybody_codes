## **Part I**

One of the grand tournament’s most time‑honoured traditions is the **Master of the Royal Forge** challenge. Knights gather at the colossal forge, where the ringing of steel echoes across the courtyard. The King’s Smith awaits, silently gesturing toward a section where wooden logs have been carefully arranged. Massive nails — each with a head the size of a human hand — protrude from the tops of the logs. A hammer lies beside each one. Every contender stands ready.

The blacksmith explains the rules of the first round. Several nails have been driven into the top of the log, each sticking out at a different height. Your task is simple in theory: **drive the nails so that all their heads form a perfectly straight line**.

You begin hammering, but it’s difficult to tell whether a nail has moved even a fraction. It seems this will be a test of endurance… until you notice something important. Each nail’s protruding length is measured precisely in whole units. And each nail head has a small black dot in its center. Looking closely at your hammer, you find a matching dot.

You swing the hammer, aligning the dots — and the nail drives **exactly one unit deeper** with each strike.  
It’s not about endurance at all. It’s about **precision**.

The blacksmith glances at you and smiles knowingly.  
You have the lengths of all the nails (_your notes_), so you can calculate the **minimum number of strikes** needed to align them perfectly.

---

## **Example**

### **Input**

```
3
4
7
8
```

The nails protrude **3**, **4**, **7**, and **8** units.

- The shortest nail is **3 units**.
- All other nails must be driven down to match this height.

### **Strikes needed**

- 4 → 3 : **1 strike**
- 7 → 3 : **4 strikes**
- 8 → 3 : **5 strikes**

### **Total**

$$
1 + 4 + 5 = 10 \text{ strikes}
$$

---

## **Your Task**

**What is the minimum number of hammer strikes needed to level all the nails?**

---


````python
import re

from test_utilities import test

tests = [
    {
        "name": "Example 1",
        "s": """
            3
            4
            7
            8
        """,
        "expected": 10,
    },
]


@test(tests=tests)
def part_I(s: str) -> int:
    nails = [int(n) for n in re.findall(r"\d+", s)]
    min_nail = min(nails)
    return sum(n - min_nail for n in nails)
````

````python
with open("../inputs/everybody_codes_e2024_q04_p1.txt") as f:
    notes1 = f.read()


print(f"Part I: {part_I(notes1)}")
````

## **Part III**

You're getting the hang of it and swiftly complete the task. The log rotates again, revealing another section filled with nails. This time, however, the nails have **additional white dots**. Inspecting your hammer once more, you notice that the **other side** of the hammerhead has a matching white dot.

Curious, you align the white dots and strike.  
To your surprise, the nail moves **up exactly one unit**.

The nails are much longer than before, so choosing the wrong target height would waste a lot of effort. To be efficient, you must determine the **optimal height** to which all nails should be driven or pulled so that the **total number of hammer strikes is minimized**.

---

## **Example**

### **Input**

```
2
4
5
6
8
```

The nails protrude **2**, **4**, **5**, **6**, and **8** units.

Suppose you choose to level them all to **5 units**:

- Nails below 5 must be **pulled up** using white‑dot strikes
  - 2 → 5 : **3 upward strikes**
  - 4 → 5 : **1 upward strike**

- Nails above 5 must be **driven down** using black‑dot strikes
  - 6 → 5 : **1 downward strike**
  - 8 → 5 : **3 downward strikes**

### **Total**

$$
3 + 1 + 1 + 3 = 8 \text{ strikes}
$$

This is the **minimum possible** for this set of nails.

---

## **Your Task**

**What is the minimum number of hammer strikes needed to level all the nails?**


````python
import re
from statistics import median

from test_utilities import test

tests = [
    {
        "name": "Example 1",
        "s": """
            2
            4
            5
            6
            8
        """,
        "expected": 8,
    },
]


@test(tests=tests)
def part_III(s: str) -> int:
    nails = [int(n) for n in re.findall(r"\d+", s)]
    median_nail = median(nails)
    return round(sum(abs(n - median_nail) for n in nails))
````

````python
with open("../inputs/everybody_codes_e2024_q04_p3.txt") as f:
    notes3 = f.read()

print(f"Part III {part_III(notes3)}")
````
