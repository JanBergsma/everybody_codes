## **Part I**

No grand tournament is complete without lively music and spirited dancing! As evening falls, crowds gather in the main square to partake in the whimsical **Pseudo‑Random Clap Dance**. Each participant receives a number from the Master of Ceremonies, displayed prominently on their chest. Then, they choose their places in one of **four columns**, standing one behind the other to form long lines.

With everyone in position, the dance begins. It unfolds in **rounds**, and each round features one person designated as the **Clapper**:

- **Round 1:** first person in column 1
- **Round 2:** first person in column 2
- **Round 3:** first person in column 3
- **Round 4:** first person in column 4
- **Round 5:** back to column 1
- …and so on.

The Clapper moves to the **front of the column to their right**, facing its members.  
If the Clapper is from column 4, they move to the front of column 1.

All members of that column extend their arms to the sides.  
The Clapper then dances around the column:

- starting on the **left side**, high‑fiving each extended hand
- upon reaching the end, switching to the **right side** and moving back up

The crowd counts each clap aloud, starting from **ONE**.  
When the shouted number matches the Clapper’s chest number, an **absorption** occurs.

### **Absorption Rules**

A Clapper can be absorbed into the column in two ways:

- If they are on the **left side**, they are absorbed **in front of** the person they are high‑fiving.
- If they are on the **right side**, they are absorbed **behind** that person.

After absorption, the **first person from each column** combines their numbers into a single large number and shouts it before the next round begins.

Just before the start, the king gives a brief speech, ending with a warning to remain vigilant.  
You decide to practice predicting the flow of the dance, so you write down all columns (_your notes_) as seen from above.

````python
from itertools import zip_longest
import re

from tabulate import tabulate

from test_utilities import test

tests = [
    {
        "name": "Example rounds 0",
        "s": """
            2 3 4 5
            3 4 5 2
            4 5 2 3
            5 2 3 4
        """,
        "rounds": 0,
        "expected": 2345,
    },
    {
        "name": "Example rounds 1",
        "s": """
            2 3 4 5
            3 4 5 2
            4 5 2 3
            5 2 3 4
        """,
        "rounds": 1,
        "expected": 3345,
    },
    {
        "name": "Example rounds 2",
        "s": """
            2 3 4 5
            3 4 5 2
            4 5 2 3
            5 2 3 4
        """,
        "rounds": 2,
        "expected": 3245,
    },
    {
        "name": "Example rounds 10",
        "s": """
            2 3 4 5
            3 4 5 2
            4 5 2 3
            5 2 3 4
        """,
        "rounds": 10,
        "expected": 2323,
    },
]


class ClapDance:
    def __init__(self, s: str) -> None:
        square = [
            [int(c) for c in re.findall(r"\d+", l)] for l in s.strip().splitlines()
        ]
        self.columns = [list(l) for l in zip(*square)]

    def dance(self, rounds: int, debug: bool = False) -> int:
        for round in range(rounds):
            self.round(round)

            if debug:
                print(
                    f"round = {round+1}, number= {int("".join(f"{c[0]}" for c in self.columns))}"
                )
                print(self)

        return int("".join(f"{c[0]}" for c in self.columns))

    def round(self, round: int) -> None:
        clapper_col = round % len(self.columns)
        clapper = self.columns[clapper_col].pop(0)

        next_col = (clapper_col + 1) % len(self.columns)

        length = len(self.columns[next_col])
        quotient, modulo = divmod(clapper - 1, length)
        insert_idx = modulo if quotient & 1 == 0 else length - modulo

        self.columns[next_col].insert(insert_idx, clapper)

    def __str__(self) -> str:
        transposed = [l for l in zip_longest(*self.columns, fillvalue=" ")]
        return tabulate(transposed)


@test(tests=tests[:])
def part_I(s: str, rounds: int = 10) -> int:
    cd = ClapDance(s)
    return cd.dance(rounds, False)
````

## **Your Task**

**What do you get if you multiply the first number shouted for the 2024th time by the total number of dance rounds?**

---


````python
from collections import defaultdict


from test_utilities import test

tests = [
    {
        "name": "Example",
        "s": """
            2 3 4 5
            6 7 8 9
        """,
        "expected": 50877075,
    },
]


class ClapDanceII(ClapDance):

    def counting_dance(self, goal: int = 2024) -> int:
        counts = defaultdict(int)
        round = 0

        while True:
            self.round(round)

            round += 1

            shout = tuple(c[0] for c in self.columns)
            counts[shout] += 1

            if counts[shout] == goal:
                return round * int("".join(f"{i}" for i in shout))


@test(tests=tests[:])
def part_II(s: str) -> int:
    cd = ClapDanceII(s)
    return cd.counting_dance()
````

If this dance were to continue **forever**, the highest number that would ever appear at the end of a round is:

## **Your Task**

**What is the highest possible number that can appear at the end of a round in the final dance?**

---


````python
from collections import defaultdict


from test_utilities import test

tests = [
    {
        "name": "Example",
        "s": """
            2 3 4 5
            6 7 8 9
        """,
        "expected": 6584,
    },
]


class ClapDanceII(ClapDance):

    def highest_number_dance(self, goal: int = 2024) -> int:
        round = 0
        seen = set()
        max_shout = (0,)

        state = tuple(tuple(c) for c in self.columns), 1
        while state not in seen:
            seen.add(state)

            shout = tuple(c[0] for c in self.columns)

            if shout > max_shout:
                max_shout = shout

            self.round(round)

            round += 1

            state = tuple(tuple(c) for c in self.columns), round % len(state)

        return int("".join(f"{i}" for i in max_shout))


@test(tests=tests[:])
def part_III(s: str) -> int:
    cd = ClapDanceII(s)
    return cd.highest_number_dance()
````
